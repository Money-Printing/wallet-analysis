```python
from fun import *
from time import sleep
from undetected_chromedriver import Chrome, ChromeOptions
# from selenium.webdriver import Chrome, ChromeOptions
from selenium.webdriver.common.by import By
from requests import get, post
from pandas import DataFrame, read_csv
from dotenv import load_dotenv
from os import getenv
load_dotenv()
blockchair_api = getenv('blockchair_api')
```

```python
chrome_options = ChromeOptions()
chrome_options.add_argument("--user-data-dir=/home/ankur/.selenium/")
chrome_options.add_argument("--profile-directory=selenium")
driver = Chrome(options=chrome_options)
```

```python
driver.get('https://bitinfocharts.com/bitcoin/wallet/Bitfinex-coldwallet')
sleep(5)
driver.execute_script("$('#ShowAddresesContainer').toggle();")
```

```python
bitfinex_btc_address = DataFrame(list(map(lambda e: e.text, driver.find_elements(By.XPATH, "//div[@id='ShowAddresesContainer']/a"))), columns=['Address'])
bitfinex_btc_address
```

```python
csv = getenv('bitfinex_btc_wallets_csv_url')
bitfinex_btc_address.to_csv(csv, index=False)
read_csv(csv)
```

```python
driver.quit()
```

```python
data = get(f'https://api.blockchair.com/bitcoin/dashboards/addresses/{",".join(bitfinex_btc_address["Address"].to_list())}?key={blockchair_api}', timeout=10).json()
data
```

```python
df = DataFrame(data['data']['addresses']).transpose().drop(columns=['type', 'script_hex', 'output_count', 'unspent_output_count']).sort_values(by=['balance'], ascending=False)
df.index.name = 'Address'
df = df.reset_index()
df.index.name = 'Ranking'
df[['balance', 'received', 'spent']] /= 1e8
df
```
